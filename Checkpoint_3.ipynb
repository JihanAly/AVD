# **Matplotlib & Seaborn**

Tema: Cultural Tourism Dataset (Analisis Tren dan Preferensi Wisatawan dalam Pariwisata Budaya)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

file = "/content/tourism_dataset_5000.csv"

df = pd.read_csv(file)
df
```

**Studi Kasus:**

1. Apakah VR bisa menjadi solusi bagi wisatawan yang mengalami kendala akses ke destinasi budaya tertentu?

2. Apakah ada hubungan antara umur dan destinasi yang dipilih?

3. Apa saja faktor yang mempengaruhi kepuasan wisatawan di destinasi budaya?

```python
1. # Bar Chart (Seaborn) - Rata-rata kepuasan wisatawan berdasarkan penggunaan VR
plt.figure(figsize=(8, 5))
sns.countplot(x='VR Experience Quality', data=df,hue="Satisfaction", palette='pastel')
plt.title('Rata-rata Kepuasan Wisatawan berdasarkan Penggunaan VR')
plt.xlabel('Penggunaan VR')
plt.ylabel('Rata-rata Kepuasan')
plt.xticks(rotation=0)
plt.show()
```

Visualisasi menunjukkan bahwa kepuasan wisatawan terhadap destinasi budaya bervariasi berdasarkan tingkat penggunaan VR, dengan kepuasan yang lebih tinggi cenderung lebih sedikit dibandingkan kepuasan yang lebih rendah. Wisatawan yang memberikan rating kepuasan rendah (3) memiliki jumlah yang jauh lebih besar dibandingkan dengan rating kepuasan tinggi (5), yang mengindikasikan bahwa penggunaan VR belum sepenuhnya meningkatkan pengalaman wisata secara signifikan.

```python
2. #BLOXPLOT â€“ hubungan antara umur dan destinasi yang dipilih
plt.figure(figsize=(12, 6))
sns.boxplot(x="Site Name", y="Age", data=df, palette="Set2")
plt.xlabel("Destinasi Budaya")
plt.ylabel("Usia Wisatawan")
plt.title("hubungan antara umur dan destinasi yang dipilih")
plt.xticks(rotation=15)
plt.show()
```

Visualisasi menunjukkan bahwa usia wisatawan yang mengunjungi berbagai destinasi budaya relatif serupa, dengan rentang usia yang luas dari sekitar 18 hingga 70 tahun. Median usia wisatawan berkisar antara 40 hingga 50 tahun, yang mengindikasikan bahwa destinasi budaya menarik minat wisatawan dari berbagai kelompok usia, terutama wisatawan dewasa. Tidak ada destinasi yang secara khusus lebih dominan dikunjungi oleh kelompok usia tertentu.

```python
3. #BLOXPLOT â€“ faktor yang mempengaruhi kepuasan wisatawan di destinasi budaya
plt.figure(figsize=(8, 5))
sns.boxplot(x='Tourist Rating', y='Recommendation Accuracy', data=df, palette='coolwarm')
plt.title('faktor yang mempengaruhi kepuasan wisatawan di destinasi budaya')
plt.xlabel('Tourist Rating')
plt.ylabel('Recommendation Accuracy')
plt.xticks(rotation=85)
plt.show()
```

Dari boxplot ini, terlihat bahwa semakin tinggi rating wisatawan terhadap destinasi budaya, cenderung lebih tinggi dan stabil. Wisatawan yang memberikan rating rendah (1.0 - 2.0) memiliki rekomendasi yang lebih bervariasi, dengan beberapa data menunjukkan akurasi yang lebih rendah dibandingkan rating tinggi. Untuk rating yang lebih tinggi (4.0 - 5.0), median akurasi rekomendasi lebih stabil dan cenderung di atas 90%, yang menunjukkan bahwa wisatawan lebih puas ketika sistem rekomendasi memberikan hasil yang lebih akurat. Rating yang lebih rendah memiliki interquartile range (IQR) yang lebih luas, artinya ada beberapa wisatawan yang merasa rekomendasinya tidak sesuai.
Pada rating 3.0 - 3.5, distribusi lebih netral, menandakan adanya ketidakkonsistenan dalam akurasi rekomendasi.
